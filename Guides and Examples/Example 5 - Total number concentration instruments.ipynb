## Example 5 - Total number concentration instruments <a class="anchor" id="TOC10"></a>

In this example we will be loading and plotting a total number concentration dataset from a condensation particle counter (CPC), so an instrument without particle size distribution data.

```python
import NFA_Aerosol.Instrument_Lib as IL
import NFA_Aerosol.Plot_Lib as PL
import NFA_Aerosol.Utility_Lib as UL

d1,header = IL.Load_CPC(r"C:\Users\B279683\Desktop\Example Data\CPC.TXT")

fig,ax = PL.Plot_totalconc(d1,y_lim = (0,8000),elapsed=1)
```

Here we simply load the dataset from the CPC file via the IL.Load_CPC() function, which only returns the data and a header, unlike other load function that return 3 outputs because they also have sizebin data and therefore return bin_egdes as well.

We then plot the dataset directly via the PL.Plot_totalconc function, and here we asked to plot it as elapsed, so that the plot starts from time zero.
