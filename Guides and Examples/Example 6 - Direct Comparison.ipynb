## Example 6 - Direct Comparison <a class="anchor" id="TOC11"></a>

Here we will be looking at a direct comparison between two of the same instruments, which have been measuring in parallel. The idea of this function is to quickly visualize the difference between the two instruments e.g. via total number concentration offset or specific particle size bin differences. The method is meant to be used before field campaigns, to determine whether a correction factor should be used when comparing the two instruments. In this example the first instrument measures concentrations, which are 1.12 times higher than the other, so a correction factor of 1.12 should be used for comparison. However, it is also clear that there is a significant disagreement for the reported concentrations of the largest particle sizes. This could call for further investigation or at least show that comparison of these size concentrations should be done with care.  

NOTE! The script is sensitive towards time allignment as it makes a direct comparison between the two datasets. You may therefore need to crop the datasets, so that they have exactly the same number of datapoints and potentially also shift the datasets in order to align identical peaks in the datasets.

```python
import NFA_Aerosol.Instrument_Lib as IL
import NFA_Aerosol.Plot_Lib as PL
import NFA_Aerosol.Utility_Lib as UL

d1, bin_edges, header = IL.Load_Nanoscan(r"Example Data\NanoScan1.csv")
d2, _, _ = IL.Load_Nanoscan(r"Example Data\NanoScan2.csv")

fig,ax = PL.Direct_compare(d1,d2,bin_edges)
```

This example will not be covered in detail as it is very short. Instead I will give a brief description of the three figures:

- Upper left plot: Total particle number concentration reported by the two instruments plotted side by side. The first dataset given to the function will always be red and the seconds will be blue.

- Lower left plot: Ratios of the total number concentrations plotted in the upper left plot (instrument1 / instrument2).  Here a 1:1 line is plotted as a red dashed line, while the mean of the ratios is plotted as a black line with its standard deviation as a black dashed line. The values of the mean and std are shown in the figure legend.

- Upper right plot: Average particle size distributions measured during the test for both instruments. This is to visualize the difference between specific size bins

- Lower right plot: Ratios of the average PSDs from the upper right plot. A 1:1 line is plotted as a red dashed line, corresponding to perfect agreement between the two instruments.
